```python
import pandas as pd
import numpy as np
import os
#Q1:Defina um diretório de trabalho 
f='C:/Users/gabri/OneDrive'
os.chdir(f)
print(os.getcwd())
```

```python
#Q2:Importe os 3 arquivos de dados para o Python

#arquivo 1
A='https://gitlab.com/lemaufpb/dspy/-/raw/main/data/dim_escola.xlsx'
d_escolas=pd.read_excel(A,sheet_name='dados',
                        names=['cod_escola','cod_municipio','cod_rede','localizacao','nome_escola','situacao_funcionamento'])
d_escolas
```

```python
#arquivo 2
B='https://gitlab.com/lemaufpb/dspy/-/raw/main/data/fato_rendimento_escolar.csv'
rend_escolar=pd.read_csv(B, sep='|',skiprows=5, na_values='-').\
rename(columns={'Ano':'Anos','Código da Escola':'cod_escola',
                             'Taxa de Abandono no Ensino Médio (%)':'Taxa_de_Abandono'})
rend_escolar['Taxa_de_Abandono'] = rend_escolar['Taxa_de_Abandono'].str.replace(',','.').astype(float)
rend_escolar
```

```python
#arquivo 3
l='C:/Users/gabri/OneDrive/prova2/RELATORIO_DTB_BRASIL_MUNICIPIO (1).xls'

colunas=['UF','Nome_UF','cod_Região_Intermediária','Nome_Região_Intermediária',
         'cod_Região_Imediata','Nome_Região_Imediata','cod_Mesorregião',
         'Nome_Mesorregião','cod_Microrregião','Nome_Microrregião',
         'Município','cod_municipio','Nome_Município' ]
rdm=pd.read_excel(l,names=colunas)
rdm['id'] =rdm.index
rdm
```

```python
# Ao final faça uma inspeção nos dados usando .info() em cada objeto.

d_escolas.info()

rend_escolar.info()

rdm.info()
```

```python
#Q3:Crie um objeto chamado df a partir da junção das tabelas dim_escola e dtb e dos seguintes filtros:
#escolas da rede estadual de ensino; em atividade; no estado da Paraíba.

p=pd.merge(rdm,d_escolas , on='cod_municipio', how='left')
df=p.query('Nome_UF=="Paraíba"').\
query('cod_rede==2').\
query('situacao_funcionamento == 1')                                                        
df
```

```python
#Q4:Junte ao objeto df, os dados da tabela fato_rendimento_escolar, filtrando o período de 2015 a 2021.

df2=pd.merge(df,rend_escolar, on='cod_escola', how='inner')
df2=df2.query('Anos!=2014').replace(np.nan,0)
df2
```

```python
#Q5:Imprima uma relação com o nome da escola, nome do município e a taxa de abandono das 20 escolas com
#maior nível de abandono escolar em 2021.
q5=df2.query('Anos==2021').\
filter(items=['nome_escola','Nome_Município','Taxa_de_Abandono']).\
sort_values('Taxa_de_Abandono', ascending= False).\
head(20)
q5
```

```python
#Q6:Imprima uma relação com o nome da escola, nome do município, nome da mesorregião e a taxa de abandono
#das escolas com o maior nível de abandono escolar em 2021 em cada microrregião do estado da Paraíba.
q6=df2.sort_values('Taxa_de_Abandono', ascending=False).\
query('Anos==2021').\
filter(items=['nome_escola','Nome_Município','Nome_Mesorregião','Taxa_de_Abandono','Nome_Microrregião'])
q6.drop_duplicates(subset=['Nome_Microrregião'])

```

```python
#Q7:Imprima uma relação com o código da escola, nome da escola, localização e a diferença da taxa de abandono
#das 10 escolas com as maiores diferenças das taxas de abandono entre os anos de 2019 (antes da pandemia)
#e 2021 (no período da pandemia).

ano_19=df2.query('Anos==2019')
ano_21=df2.query('Anos==2021')
td=ano_19.merge(ano_21,on=['cod_escola','nome_escola','localizacao'])
q7=td.filter(items=['cod_escola','nome_escola','localizacao'])
q7['diff_2019|2021']=td['Taxa_de_Abandono_y']-td['Taxa_de_Abandono_x']
q7.sort_values('diff_2019|2021', ascending=False, inplace=True)

q7.head(10)
```

Q11: Com base nas análises de dados realizadas, o que é possível concluir sobre os efeitos da pandemia no nível
de abandono escolar no ensino médio?


 Com base nos dados sobre a taxa de abandono das escolas estaduais da paraíba, pode-se concluir que a taxa de abandono escolar de modo geral vem diminuindo ao longo dos anos  ,vemos esse fenômeno ao comparar as taxas de 2015 com as de 2021 que são respectivamente de 13.63% e 3.07% ,contudo, essa diminuição não é linear por ter  aumentos nos anos 2016 e 2021, mas ao comparar a taxa dos anos inicial e final dos dados temos uma queda de mais de 10%.
 No âmago pré e pós pandemia houve uma diminuição de quase 50% na taxa de abandono ao comparar 2019 com 6.95% e 2021 com 3.07%,já no período pandêmico de 2020 a 2021 houve um aumento a taxa era de 2,07% e passou a ser de 3,07% .  
 Nas escolas estaduais de localização urbanas da Paraíba temos um  aumento de 2015 para 2016 e  uma diminuição linear de 2016 à 2020 e um aumento em 2021. O mesmo fenômeno acontece nas escolas de localização Rural da Paraíba .Ao comparar a taxa de abandono das escolas urbanas  com as escolas rurais vemos que  em todo o  período da pesquisa os dados demonstram que as escolas urbanas sofrem em média mais abandonos  que as escolas rurais . 
 As mesorregiões que mais sofrem com o abandono escolar são o Agreste paraibano e o Sertão paraibano ,as escolas dessas duas mesorregiões ocupam 15 das 23  escolas com o maior nível de abandono escolar em 2021.
 
 O efeito da Pandemia nas escolas estaduais de ensino médio da Paraíba na  taxa de abandono escolar foi uma oscilação por ter uma queda significativa de 2019 para 2020  e um aumento em 2021  que não superam os valores de antes da pandemia, então pode-se concluir que até o ano de 2021 a pandemia não afetou excessivamente as taxas de abandono mas hove uma queda considerável.
